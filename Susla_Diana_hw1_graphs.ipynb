```
import sys

OK = '\033[92m'
WRONG = '\033[91m'
END = '\033[0m'

input_file = '/content/drive/My Drive/tests/01[1].in'
output_file = '/content/drive/My Drive/tests/my,out'
answer_file = '/content/drive/My Drive/tests/01[1].ans'

with open(input_file, 'r') as f:
    n, m = [int(x) for x in next(f).split()]
    edges = []
    for _ in range(m):
        u, v = [int(x) for x in next(f).split()]
        edges.append((u - 1, v - 1))

with open(output_file, 'w') as f:
    def transform(edges, n, k):
        def get_inx(v, j):  #return (v + 1, j + 1)
            return (v + 1) * k + j + 1
        clauses = []
    # xi1 or ... or xin = 1, данное условие гарантирует, что у каждой вершины есть хотя бы какой-то цвет
        for v in range(n):
            clause = []
            for i in range(k):
                clause.append(get_inx(v, i))
            clauses.append(clause)
    # not (xij and xik), i != k, данное условие исключает раскраску одной веришины в несколько цветов
        for v in range(n):
            for i in range(k):
                for j in range(i + 1, k):
                    clause = [-get_inx(v, i), -get_inx(v, j)]
                    clauses.append(clause)
    # not (xvj and xuj), (u, v) in edges, данное условие исключает наличие одинаковой раскраски вершин, между которыми есть ребро
        for u, v in edges:
            for i in range(k):
                clause = [-get_inx(u, i), -get_inx(v, i)]
                clauses.append(clause)

    #for u in range(n):
    #    for v in range(u + 1, n):
    #        if (matrix[u][v] != 0):
    #            for i in range(k):
    #                clause = [-get_inx(u, i), -get_inx(v, i)]
    #                clauses.append(clause)
    
        return clauses
    i = 1
    chromatic = 0
    from pysat import solvers

    while i <= 5:
        clauses = transform(edges, n, i)
        solver = solvers.Glucose3()
        for clause in clauses:
            solver.add_clause(clause)
        if (not solver.solve()):
            i += 1
        else:
            chromatic = i
            break
    f.write(str(chromatic))
    f.write('\n')
   
    for j in solver.get_model(): #пусть цвет - остаток от деления числа j - индекса вершины, на общее количество цветов (хроматическое число)
        if j > 0 :
            if (j % i == 0):
                f.write(str(i))
                f.write(' ')
            else:
                f.write(str(j % i))
                f.write(' ')

with open(output_file, 'r') as f:
    chromatic_number = int(next(f))
    coloring = [int(x) for x in next(f).split()]
    if len(coloring) != n:
        print(WRONG, 'Wrong:', END, 'expected {} colors, got {}'.format(n, len(coloring)))
        sys.exit()

    for index, color in enumerate(coloring):
        if color <= 0:
            print(WRONG, 'Wrong:', END, 'c[{}] <= 0'.format(index + 1))
            sys.exit()
        if color > chromatic_number:
            print(WRONG, 'Wrong:', END, 'c[{}] > {}'.format(index + 1, chromatic_number))
            sys.exit()

for u, v in edges:
    if coloring[u] == coloring[v]:
        print(WRONG, 'Wrong:', END, 'vertices {} and {} are connected by an edge but colored with the same color'.format(u + 1, v + 1))
        sys.exit()

if answer_file is not None:
    with open(answer_file, 'r') as f:
        answer = int(next(f))
        if chromatic_number != answer:
            print(WRONG, 'Wrong:', END, 'your chromatic number is {} but the real chromatic number is {}'.format(chromatic_number, answer))
            sys.exit()

print("Первый тест:")
print(OK + 'Ok!' + END)
```

```
import time
start_time = time.time()

import networkx as nx
def transform(matrix, n, k):
    def get_inx(v, j):
        return (v + 1) * k + j + 1
    clauses = []
    for v in range(n):
        clause = []
        for i in range(k):
            clause.append(get_inx(v, i))
        clauses.append(clause)
    for v in range(n):
        for i in range(k):
            for j in range(i + 1, k):
                clause = [-get_inx(v, i), -get_inx(v, j)]
                clauses.append(clause)
    for u, v in matrix:
        for i in range(k):
            clause = [-get_inx(u, i), -get_inx(v, i)]
            clauses.append(clause)
            
    return clauses



G = nx.gnm_random_graph(600, 456)
matrix = G.edges

#for u, v in matrix:
#    print(u, " ", v)
    
n = 600
i = 1
chromatic = 0
from pysat import solvers

while i <= n:
    clauses = transform(matrix, n, i)
    solver = solvers.Glucose3()
    for clause in clauses:
        solver.add_clause(clause)
    if (not solver.solve()):
        i += 1
    else:
        chromatic = i
        break

print("Хроматическое число = ", chromatic)

ans = []
for j in solver.get_model():
    if j > 0:
        if (j % i == 0):
            ans.append(i)
        else:
            ans.append(j % i)
print("Правильная раскраска графа:")
print(ans)

print("Время выполнения алгоритма составляет:")
print("%s seconds" % (time.time() - start_time))
```
